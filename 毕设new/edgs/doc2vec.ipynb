# doc2vec普通方法

```python
#doc2vec

import jieba 
import gensim
from gensim.models import doc2vec

from MyDataFrame import MyDataFrame
mdf = MyDataFrame()
df = mdf.new_DataFrame()
df2 = mdf.m_cut(df)

def m_doc(doclist):
    reslist = []
    for i, doc in enumerate(doclist):
        reslist.append(doc2vec.TaggedDocument(doc, [i]))
    return reslist

corp = m_doc(df2.fenci)
print(corp[:2])

d2vmodel = gensim.models.Doc2Vec(vector_size = 100, window = 8, min_count = 5,workers=8)
%time d2vmodel.build_vocab(corp) #52s
d2vmodel.wv.vocab #dict

# 将新文本转换为相应维度空间下的向量
newvec = d2vmodel.infer_vector(df2.fenci[0])
d2vmodel.docvecs.most_similar([newvec], topn = 10)
```
