# 分词并去除停用词

```python
import jieba
import pandas as pd
dict='C:/TMdata/计算机词汇.txt'
jieba.load_userdict(dict)
stoplist = list(pd.read_csv('C:/TMdata/stopwords.txt',names=['w'],sep='aaa',encoding='UTF-8',engine='python').w)
#print(stoplist)
f=open(r'C:\EdmsData\txt\大数据时代的计算机网络安全及防范技术_欧东鑫.txt','r',encoding='utf-8',errors='ignore')
str1 = ''
for line in f.readlines():
    a = line.strip('\n')
    str1 = str1 + a
listi = [w for w in jieba.cut(str1) if w not in stoplist and len(w) >1]
print(listi)
print(len(listi))
```
